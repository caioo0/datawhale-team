 ### 1. 函数
 
**函数定义**
 
 
 通俗地讲，**函数**就是数与数之间对应关系的抽象代数化。最简单的就是一元函数，如$y=x$，其中每一个$x$都唯一对应一个$y$。函数种类繁多，常见的一般有：`绝对值函数`，`符号函数`，`取整函数`，`狄里克雷函数`。
 
 每一个函数都会在一定程度上表现出某些共性，所以就有了以下的特征概念：
 
 - 单调性（在某区间内函数值恒增或恒减）
 - 奇偶性（f(x)=f(-x)/f(x)=-f(-x))
 - 周期性（等区间上函数波形的复现
 - 有界性 

 
有些时候，因为运算的需要，还用到一些常用的函数形式：`反函数`、`隐函数`、`参数函数`、`复合函数`。

### 2. 导数

正式讲导数之前，我们首先从函数的连续性开始讨论，然后逐步过渡到切线和导数的概念。


**函数的连续性** 


函数 $f(x)$ 在具体的取值点 $c$ 点是否连续，我们先来看下面示例图（图1）：


![cp01_img01.png](images/cp01_img01.png) 

$$(图1)$$
在这幅图中，我们发现在点 $c$ 处，函数的左极限和右极限不相等，即

$$
\begin{aligned} \lim _{x \rightarrow c^{-}} f(x) \neq \lim _{x \rightarrow c^{+}} f(x) \end{aligned}
$$

因此 $
\begin{aligned} \lim _{x \rightarrow c} f(x)  \end{aligned}
$
 不存在，$c$ 点处函数的极限不存在，因此函数 $f(x)$ 在点 $c$ 处**不连续**。

我们来看另一种情况（图2）：
![cp01_img02.png](images/cp01_img02.png)  

$$(图2)$$

这幅图中情况似乎要稍微好点儿，我们发现

$$
\begin{aligned} \lim _{x \rightarrow c^{-}} f(x) = \lim _{x \rightarrow c^{+}} f(x) \end{aligned}
$$

函数 $f(x)$ 在 $c$ 点处的极限是存在的，但是从图中可以看出，函数的极限值和 $c$ 点处函数的实际取值不相等，即：

$$
\begin{aligned} \lim _{x \rightarrow c } f(x) \neq f(c) \end{aligned}
$$

因此函数 $f(x)$ 在点 $c$ 处仍然**不连续**。

再来看看下面一种情况（图3）：

![cp01_img03.png](images/cp01_img03.png) 


$$(图3)$$

这幅图中以上出现的两个问题都不存在了，我们看到：

- 一方面函数 f(x) 的极限是存在的
- 而另一方面$
\begin{aligned} \lim _{x \rightarrow c } f(x) = f(c) \end{aligned}
$
 点 c 处的极限和函数的取值又是相等的。
 
 因此在这幅图中，函数$f(x)$ 在点 $c$ 处是**连续**的。

> **连续性的概念**：    
  连续的定义是基于极限而来的，即一点处的值等于该点的左右极限的值。  
  同理可分为右连续和左连续
  那么不连续呢？则成为间断。   
  间断分第一类间断（可去间断点和跳跃间断点）和第二类间断（无穷间断点）。  
  连续函数的性质：闭区间上存在最大最小值；介值定理。  
  
  
  
**函数在闭区间上[a,b]的连续性**

如果函数 $f(x)$ 在开区间上的任意一点连续，那么这个函数 $f(x)$ 就在整个这个开区间上连续。


如果谈到闭区间 [a,b] 上的连续性问题，那么就需要着重单独讨论区间的左右两个端点：我们首先从右侧逼近左侧端点 a，如果 

$$
\begin{aligned} \lim _{x \rightarrow a^{+}} f(x) =  f(a) \end{aligned}
$$

 成立，则称函数 f(x) 在端点 a 上右连续，我们再从左侧逼近右侧端点 b，即 
$$
\begin{aligned} \lim _{x \rightarrow a^{-}} f(x) =  f(a) \end{aligned}
$$


 成立，则称函数 f(x) 在端点 b 上右连续。

那么，如果函数在开区间 $(a,b)$ 上连续，且在左侧端点 $a$ 上右连续，在右侧端点 $b$ 上左连续，此时此刻，我们就能够说函数 $f(x)$ 在闭区间 [a,b] 上是连续的。
